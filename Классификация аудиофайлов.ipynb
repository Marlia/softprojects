```python
import librosa
import librosa.display
import numpy as np
import pandas as pd
import IPython
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Создадим списки, которые будут хранить соответствующие характеристики аудиофайлов
content = os.listdir('genres')
names = []
zrate = []
zrate_std = []
mfcc_m = []
mfcc_m_std = []
sp_rolloff = []
sp_rolloff_std = []
rms_std = []
sp_flatness = []
sp_flatness_std = []
sp_cent = []
sp_cent_std = []
genre = []
tempo = []
rms = []

#Функция для извлечения характеристик аудиофайлов в списки (Которые позже объединятся в датасет)
def extr_f(audio_file_name,g,i):
    y, sr = librosa.load(audio_file_name)
    names.append(g+str(i))
    mfcc_m.append(np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=10), axis=1))
    
    zrate.append(float(np.mean(librosa.feature.zero_crossing_rate(y), axis=1)))
    zrate_std.append(float(np.std(librosa.feature.zero_crossing_rate(y), axis=1))) #
    
    onset_env = librosa.onset.onset_strength(y=y, sr=sr)
    
    tempo.append(float(librosa.feature.tempo(onset_envelope=onset_env, sr=sr))) 

    sp_cent.append(float(np.mean(librosa.feature.spectral_centroid(y=y, sr=sr), axis=1)))
    sp_cent_std.append(float(np.std(librosa.feature.spectral_centroid(y=y, sr=sr), axis=1))) #

    sp_flatness.append(float(np.mean(librosa.feature.spectral_flatness(y=y), axis=1)))
    sp_flatness_std.append(float(np.std(librosa.feature.spectral_flatness(y=y), axis=1)))#
    
    sp_rolloff.append(float(np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr), axis=1)))
    sp_rolloff_std.append(float(np.std(librosa.feature.spectral_rolloff(y=y, sr=sr), axis=1)))#
    
    rms.append(float(np.mean(librosa.feature.rms(y=y), axis=1)))
    rms_std.append(float(np.std(librosa.feature.rms(y=y), axis=1)))
    
    genre.append(g)
    print(f"{audio_file_name} is done")
```

```python
#Загрузка данных 
for g in content:
    for i in range(0, 100):
        if i < 10:
            i = f"0{i}"
        extr_f(str("genres/"+g+ "/" + g+ ".000"+str(i)+".au"), g,i)
```

```python
#Создание датафрейма
labels = ["MFCC_" + str(i) for i in range(1,11)]
df = pd.DataFrame(mfcc_m, columns=labels)
df.index = names

df["ZCR_mean"] = zrate
df["ZCR_std"] = zrate_std
df["sp_rolloff_mean"]=sp_rolloff
df["sp_rolloff_std"] = sp_rolloff_std
df["rms_mean"]= rms
df["rms_std"]=rms_std
df["sp_flatness"] = sp_flatness
df["sp_flat_std"] = sp_flatness_std
df["tempo"] = tempo
df["sp_centroid"] = sp_cent
df["sp_cent_std"] = sp_cent_std
df["genre"]=genre
```

```python
#Просмотр созданного датасета 
df.head()
df.describe()
```

```python
#Разделим датасет на х и у
X = df.drop("genre", axis=1)
y = df["genre"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
from sklearn.preprocessing import normalize
scaled_X_train = normalize(X_train)
scaled_X_test = normalize(X_test)

```

```python
#Создадим модель случайного леса
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200, max_features="sqrt", bootstrap=True)
rfc.fit(scaled_X_train, y_train)
```

```python
#Предскажем значения на тестовой выборке
pred = rfc.predict(scaled_X_test)
```

```python
#Посмотрим отчёт о точности модели.
from sklearn.metrics import classification_report
print(classification_report(pred, y_test))
```

```python
#Обучим модель на всём датасете
rfc_final = RandomForestClassifier(n_estimators=100, max_features="sqrt")
rfc_final.fit(X, y)
```

```python
#Сохраним обученную модель
from joblib import dump
dump(rfc_final, "final_model.joblib")
```
